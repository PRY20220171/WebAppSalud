```python
#Correccion de valores:
import random
import itertools
values=["P","M","H","A","O"]
ff = open('db.json',mode='r')
fo = open('auxdb.json',mode='w')
faux = open('antecedentefamdb.json',mode='w')
flag=False
subflag=False
cont = 0
for line in ff:
    aux = line
    if not 'idantecedentefam' in line and 'antecedentefam' in line:
        flag = True
        subflag = True
    if flag:
        if "alcoholismo" in aux:
            cantfam = random.randint(1,len(values)-1)
            temp = list(itertools.combinations(values,cantfam))
            aux = aux.replace("\"No sabe\"",
                               "[]")
            aux = aux.replace("\"Un familiar sufre de alcoholismo\"",
                               str(list(temp[random.randint(0,len(temp)-1)])).replace("'","\""))
            aux = aux.replace("\"No hay familiar con alcoholismo\"",
                               "[]")
            flag=False
            cont+=1
    if subflag:
        if not 'idantecedentepato' in line:
            faux.write(aux.split('"antecedentefam": ')[-1])
        else:
            subflag = False
    fo.write(aux)

print(cont)
fo.close()
faux.close()
```
