```python
#Correccion de valores:
import random
values=["A","P","H"]
ff = open('db.json',mode='r')
fo = open('auxdb.json',mode='w')
faux = open('antecedentefamdb.json',mode='w')
flag=False
subflag=False
cont = 0
for line in ff:
    aux = line
    if not 'idantecedentefam' in line and 'antecedentefam' in line:
        flag = True
        subflag = True
    if flag:
        if "Un familiar sufre de alcoholismo" in aux:
            aux = line.replace("Un familiar sufre de alcoholismo",
                               values[random.randint(0,len(values)-1)])
            flag=False
            cont+=1
    if subflag:
        if not 'idantecedentepato' in line:
            faux.write(aux.split('"antecedentefam": ')[-1])
        else:
            subflag = False
    fo.write(aux)

print(cont)
fo.close()
faux.close()
```
